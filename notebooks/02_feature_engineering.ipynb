```python
import pandas as pd
import os
df = pd.read_csv('data/processed/cleaned.csv')
df = df.sort_values('DIM_TIME')
def build_features(sub):
    sub = sub.copy()
    sub['rate'] = sub['RATE_PER_100_N']
    sub['yoy_change'] = sub['rate'].pct_change()
    sub['rolling_mean_3'] = sub['rate'].rolling(3).mean()
    return sub[['DIM_TIME','Sex','rate','yoy_change','rolling_mean_3']]
features = df.groupby('Sex').apply(build_features).reset_index(drop=True)
os.makedirs('data/processed', exist_ok=True)
features.to_csv('data/processed/features.csv', index=False)
features.head()
```
