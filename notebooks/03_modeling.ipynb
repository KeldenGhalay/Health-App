```python
import pandas as pd
import os
import matplotlib.pyplot as plt
path = 'data/processed/cleaned.csv'
assert os.path.exists(path), 'Run 01_data_exploration.ipynb first'
df = pd.read_csv(path)
assert {'DIM_TIME','Sex','RATE_PER_100_N'}.issubset(df.columns), 'Required columns missing'
pivot = df.pivot_table(index='DIM_TIME', columns='Sex', values='RATE_PER_100_N')
pivot.head()
```

```python
# Gender gap and change KPIs
first_year = pivot.index.min()
last_year = pivot.index.max()
kpis = {}
if 'Total' in pivot.columns:
    kpis['delta_total'] = pivot.loc[last_year,'Total'] - pivot.loc[first_year,'Total']
if {'Male','Female'}.issubset(pivot.columns):
    kpis['avg_gap_male_female'] = (pivot['Male'] - pivot['Female']).mean()
kpis
```

```python
# Plot trends with optional uncertainty bands if available
plt.figure(figsize=(8,5))
for col in pivot.columns:
    plt.plot(pivot.index, pivot[col], label=col)
plt.xlabel('Year')
plt.ylabel('Rate per 100 (ages 30â€“70)')
plt.legend()
plt.show()
```
